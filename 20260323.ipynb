```python
var1 = int(input())
var2 = int(input())
op = input()
var3 = 0

if op == '+':
    var3 = var1 + var2
if op == '-':
    var3 = var1 - var2
if op == '*':
    var3 = var1 * var2
if op == '/':
    var3 = var1 / var2


print(f"{var1} {op} {var2} = {var3}")
```
