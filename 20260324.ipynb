```python
marks = [90, 25, 90, 45, 80]

for i in range(0, len(marks)):
    if marks[i] >= 60:
        print(f"{i+1}번 학생 합격")
    else:
        print(f"{i+1}번 학생 불합격")
```

```python
print("끝말잇기 게임 시작")
var = "가나다"
next = ""
print(var)
while True:
    next = input()
    if(var[-1] == next[0]):
        var = next
        print(var)
        continue
    else:
        print("틀림")
        break
```

```python
dd = dict()
while True:
    menu = int(input("메뉴 선택 1:추가, 2:검색, 0:종료"))
    if menu == 0:
        print("종료합니다.")
        break
    if menu == 1:
        name = input("이름을 입력하세요:")
        num = input("전화번호를 입력하세요:")
        dd[name] = num
    if menu == 2:
        find_name = input("검색하실 이름을 입력하세요:")
        dd.get(find_name)
        
        
```
