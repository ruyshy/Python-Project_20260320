```python
score1 = [[80, 70], [90, 50], [40, 70], [50, 80]]
score2 = [[80, 70], [70, 80], [30, 50], [90, 100], [100, 90], [100, 100], [10, 30]]

def func1(x):
    result = [sum(i) / len(i) for i in x]
    sort_result = sorted(enumerate(result, 1), key=lambda x: x[1], reverse=True)

    ranks = {}
    current_rank = 0
    prev_score = None

    for idx, (num, avg) in enumerate(sort_result, 1):
        if avg != prev_score:
            current_rank = idx
        ranks[num] = current_rank
        prev_score = avg

    return [ranks[i] for i in range(1, len(x) + 1)]

print(func1(score1))
print(func1(score2))
```

```python
class Person:
    population = 0
    def __init__(self, name, age):
        self.name = name
        self.age = age
        Person.population += 1
        
    def walk(self):
        print(f"{self.name}이 걸어가요.")

class NewPerson(Person):
    def fly(self):
        print(f"{self.name}이 날아가요.")
    def walk(self):
        print(f"{self.name}이 빠르게 걸어가요.")
    def swim(self):
        print(f"{self.name}이 수영해요.")

person1 = Person("jimin", 31)
person1.walk()
person2 = NewPerson("jongcock", 34)
person2.walk()
person2.fly()
person2.swim()
```
